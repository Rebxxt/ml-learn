```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(10)
dataset = rng.random((100, 1), np.float64) * 15 - 4
X = dataset[:, 0]
rng2 = np.random.default_rng(15)
Y = rng2.random((100), np.float64) * 2 - 1
Y = ((2.5 * X[:] + 3) + Y[:])
plt.plot(X, Y, linestyle='', marker='.')
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
X_train = X.reshape(-1, 1)[:-20]
X_test = X.reshape(-1, 1)[-20:]
Y_train = Y.reshape(-1, 1)[:-20]
Y_test = Y.reshape(-1, 1)[-20:]
print(X_train.shape, Y_train.shape)
regr.fit(X_train, Y_train)
Y_pred = regr.predict(X_test)
plt.plot(X_test, Y_pred, marker='.')
plt.scatter(X_test, Y_test)
plt.show()
```
